```python
# import the necessary libraries and to reduce typing give the an alias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline  
# The percent symbol before matplotlib inline
#I believe, if I can still remember, helps to plot the plt graph within the current notebook.
```

```python
# created lists of numbers with numpy array method
a = np.array([25, 50, 13, 8, 15, 30, 10])
b = np.array([5, 20, 45, 12, 10, 35, 15])
print(a.sum())
print(b.sum())
```

```python
# With the sum function, performed a matrix addition operation on the two arrays
# Assigned it to a vaiable named ab

ab = sum(a,b)
ba = sum(b,a)
print(ab, ba)
```

```python
# Performed a  subtraction operation on the arrays with the numpy substraction method
c = np.subtract(a,b)
print(c)
```

```python
# Created lists or arrays with numpy arange method and assigned them a new variable
a_list = np.arange(0, 24, 2)
b_list = np.arange(1, 25, 2)
c_list = np.array([[a_list], [b_list]])
print(c_list)
c_list.shape
```

```python
# Performed a slicing operations on the list while assigning it a new variable
a_1 = a_list[:6]
a_2 = b_list[6:]
a_3 = np.array([[a_1], [a_2]])
print(a_3)
```

```python
# Working with MATPLOTLIB

income_rate = [10.4, 15.5, 12.3, 20, 35.1, 25, 17.2, 40, 50, 60, 57.5, 38.9]
expenditure = [6.5, 10, 6, 9, 10, 7.5, 8.9, 9.9, 10, 6.5, 7.8, 5.5]
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sept", "Oct", "Nov", "Dec"]
```

```python
plt.rcParams['font.size'] = 24
plt.figure(figsize = (24, 10))
plt.plot(months, income_rate, marker = 'o', ms = 14)
plt.plot(months, expenditure, marker = 'o', ms = 14)
plt.title('Sales Report', fontsize = 24)
plt.legend(['Income', 'Expenses'], fontsize = 14)
plt.xlabel('Months', fontsize = 24)
plt.ylabel('Income&Expenditure in Thousand', fontsize = 24);
```
